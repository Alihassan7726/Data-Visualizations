```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df
```

```python
# sns.lmplot also allows 'scatter_kws' to define alpha and size .
sns.regplot(x='tip',y='total_bill', data=df, scatter_kws={'alpha': 0.5}, fit_reg=True) #fit_reg is responsible for line presenting
# relationship among both
```

```python
# Jointplot shows the density of data and distribution at the same time
sns.jointplot(x='tip',y='total_bill', data=df,)
```

```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='kde')
```

```python
sns.relplot(x='tip',y='total_bill', data=df,col='time', hue= 'time', height=3, kind='line', legend='brief')
```

```python
df['time'] = df['time'].replace(['Dinner','Lunch'],['Dinner-T','Lunch-T'])
df.head(3)
```

```python
sns.lmplot(x='tip',y='total_bill', data=df, col='time', hue='time', palette=['b','r'], ) #ci=None to remove the efects of line
```

```python
# Great for multi plotting grids
sns.FacetGrid(df,col='time').map(plt.scatter, 'tip','total_bill', alpha= 0.4).add_legend()
```

```python
sns.catplot(x='day',y='total_bill', data=df, kind='violin')
```
