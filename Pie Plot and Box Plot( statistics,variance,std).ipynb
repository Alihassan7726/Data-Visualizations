# Pie Plots

```python
import matplotlib.pyplot as plt 
# Pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [5, 10, 10, 5]
# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0, 0.1, 0, 0)  
fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels,pctdistance = 0.6, autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout(pad = 0)
plt.show()
```

```python
# Wedgeprops = {dic} is responsible for donut shape
plt.figure(figsize=(5,6))
location = ['Maharastra', 'Gujarat', 'Tamilnadu', 'Delhi', 'Rajasthan', 'Madhya Pradesh', 'Uttar pradesh']
Confirmed_cases = [23000,8500, 8000,7200,4000, 3700, 3500]

plt.style.use('tableau-colorblind10')
plt.axis('equal')

plt.pie(Confirmed_cases, labels= location, explode=[0,0.1,0,0,0,0,0],radius= 1.5, autopct= '%2.1f%%', pctdistance = 0.85,
        shadow = True, labeldistance= 1.1, counterclock= True, wedgeprops={'width': 1.1})
plt.title("Positive case from 1st may 2020 to 11th may 2020")
plt.show()
```

```python
# Creating a box plot
import numpy as np
import matplotlib.pyplot as plt
data = [1,2,3,4,5,6,7,8,16]
percent = np.percentile(data,[25,50,75])
percent
# Min is 1 and Max is 16 , but 16 is outlier so , we cannot use it instead we use 8.
```

```python
Q3 = percent[2]
Q1 = percent[0]
Q3
```

```python
IQR = Q3-Q1
whis_min = Q1 - 1.5*IQR
whis_max = Q3 + 1.5*IQR
print(IQR)
print(whis_min)
print(whis_max)
```

```python
# "whis" increasing it value to 2.5 means increaing line of max no. to outlier ,
# "vert" is used to display in hori or verti order
# "widths" is used to display width of bar
# "Showmeans" to show mean value with triangle
# "showcaps" not a great efeect on output
# "meanline" instead of trianlge shows the line of mean
plt.boxplot(data, vert = False, widths = 0.15, whis = 1.5, showmeans = True, showcaps = True, meanline = True)
```
