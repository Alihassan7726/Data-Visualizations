```python
# This is a simple graph
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
x = [1,2,3,7,10]
#y = [4,6,8,10,12]
plt.figure(figsize = (14,4))
plt.style.use('classic')
date = pd.date_range('20200501',periods = 5)
plt.plot(date,x,label = 'cases') # x and y must be of same length

plt.title("This is title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

```python
# A bit complex

from matplotlib import style
plt.figure(figsize = (8,6))
style.use('classic')
x1 = [5,8,10,11]
y1 = [4,6,8,10]

x2 = [3,6,7,9,12]
y2 = [1,3,5,7,9]
plt.plot(x1,y1,'g',label = 'line-1',linewidth = 5, alpha = 0.3) # Alpha is transparency
plt.plot(x2,y2,'m',label = 'line-2',linewidth = 5, linestyle = '--')
plt.plot(x1,y1,'b^')
plt.axis([0,15,0,13])
plt.title('Multi-line graph')
plt.xlabel('X-axis', fontsize=20, color = 'b')
plt.ylabel('Y-axis')
plt.legend(loc = 'lower right') # Display legend in lower right corner
plt.grid(True,color = 'k') # 'k' means black color
plt.show()
```

```python
# Plotting same above graph in one line
from matplotlib import style
plt.figure(figsize = (8,6))
style.use('classic')
x1 = [5,8,10,11]
y1 = [4,6,8,10]

x2 = [3,6,7,9,12]
y2 = [1,3,5,7,9]
plt.plot(x1,y1,'g',x2,y2,'m',x1,y1,'b^',linewidth = 5,label = 'line-1'"""linewidth = 5, alpha = 0.3""") # Alpha is transparency
#plt.plot(x2,y2,'m',label = 'line-2',linewidth = 5, linestyle = '--')
#plt.plot(x1,y1,'b^')
plt.axis([0,15,0,13])
plt.title('Multi-line graph')
plt.xlabel('X-axis', fontsize=20, color = 'b')
plt.ylabel('Y-axis')
plt.legend(loc = 'lower right') # Display legend in lower right corner
plt.grid(True,color = 'k') # 'k' means black color
plt.show()
```

```python
import numpy as np
# The example below illustrates plotting several lines with different format styles in one function call using arrays.
t = np.arange(0., 5., 0.2) # 25 values
# red dashes , blue squares , green triangles
# you can change 'r--'bs'g^' to 'g--'gs'b^' to get same shape in different colors --,s,^ represents shapes
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^') # it will automatically adjust the axis
plt.show()
```

```python
f, ax = plt.subplots(1,1) # Not a great effect , may it will be useful I don't know
x1 = [5,8,10,11]
y1 = [4,6,8,10]

x2 = [3,6,7,9,12]
y2 = [1,3,5,7,9]
style.use('classic')
plt.bar(x1,y1, label = 'Example-1')
plt.bar(x2,y2, color ='g', label = 'Example-2') # Note: 'g' must be a keyword argument to differentiate both
plt.legend()
plt.title('This is title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Same as above but with seaborn
import seaborn as sns
sns.set(style='dark') # Possibel inputs = ['darkgrid','whitegrid','dark','white','ticks']
f, ax = plt.subplots(1,1) # Not a great effect , may it will be useful I don't know
x1 = [5,8,10,11]
y1 = [4,6,8,10]

x2 = [3,6,7,9,12]
y2 = [1,3,5,7,9]
#style.use('classic')
plt.bar(x1,y1, label = 'Example-1')
plt.bar(x2,y2, color ='g', label = 'Example-2') # Note: 'g' must be a keyword argument to differentiate both
plt.legend()
plt.title('This is title')
ax.set(xlabel='Rating', ylabel='Frequency') # An convinient method to set axis labels
plt.show()
```

```python
# Plotting with categorical variables
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(10, 3)) # width , height (order of values)

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting') # add a centric title to figure
plt.show()
```
