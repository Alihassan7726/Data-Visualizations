# What is matplotlib? 

- Matplotlib is a plotting library for python programming language and it was inatially written by john Hunter.

# Advantages of matplotlib

* Simple and easy for beginners.
* It has large community support and cross platform support as it is an open source tool.
* It has full control over graphs or plot styles.
* It can also be used in python scripts, shells, web applications servers and other graphical user interface toolkits.
* It is fast and efficient because it is builded on numpy and scipy framework.
* It is platform - independent and can be run on windows and macos and linux 

# What are type of plots in matplotlib in python?

* Line plot - very useful when it comes to multivariate analysis.
* Histograms - to understand the distribution.
* Bar charts - we can use to compare one over the other. 
* Pie charts - to present the data in percentage, ratio, proportion etc.
* Scatterplots - to identify the relationship between two variables. 
* Subplots - Very useful when comparing two or more plots 
* Images - matplotlib can also create image with the help of imshow() funciton

# Arguments for Pie Chart

```python
explode=[0.1,0,0,0,0,0,0] #To explode specific slice

radius= 1.5 # size of the plot

autopct= '%2.1f%%' #percentage of each slice

pctdistance = 0.6 #percentage distance from the center

shadow = True #Draw a shadow beneath the pie

labeldistance= 1.1 #The radial distance at which the pie labels are drawn

counterclock= True # Specify fractions direction, clockwise or counterclockwise
```

# pie chart

```python
plt.figure(figsize=(5,6))
location = ['Maharastra', 'Gujarat', 'Tamilnadu', 'Delhi', 'Rajasthan', 'Madhya Pradesh', 'Uttar pradesh']
Confirmed_cases = [23000,8500, 8000,7200,4000, 3700, 3500]

plt.style.use('tableau-colorblind10')
plt.axis('equal')

plt.pie(Confirmed_cases, labels= location, explode=[0,0.1,0,0,0,0,0],radius= 1.5, autopct= '%2.1f%%', pctdistance = 0.85,
        shadow = True, labeldistance= 1.1, counterclock= True, wedgeprops={'width': 1.1})
plt.title("Positive case from 1st may 2020 to 11th may 2020")
plt.show()
```

# Histogram

```python
plt.figure(figsize=(8,6))
plt.style.use('tableau-colorblind10')
day_wise_cases = [1755,2411,2487,2573,3875,2680, 3561,3390,3320,3277,4213]

plt.hist(day_wise_cases,range=[1000,5000], density=False, bins= 5, rwidth=0.95,
         histtype='barstacked', label= 'day_wise_cases', color= 'grey' , orientation = 'vertical')
plt.title('Day-wise-count')
plt.xlabel('Number of cases')
plt.ylabel('counts')
plt.legend()
```
